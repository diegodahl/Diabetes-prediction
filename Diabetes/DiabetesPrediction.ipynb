```python
import pandas as pd, matplotlib.pyplot as plt
from tensorflow import keras
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras import Sequential
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age',]],
                                                    df[['Outcome']],
                                                    test_size=0.2,random_state=42)
```

```python
X_train_scaled = X_train.copy()
X_train_scaled = StandardScaler().fit_transform(X_train_scaled)

X_test_scaled = X_test.copy()
X_test_scaled = StandardScaler().fit_transform(X_test_scaled)
```

```python
#Inicializamos el modelo
model = keras.Sequential([])
model.add(keras.layers.Dense(1,input_shape=(8,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(X_train_scaled,y_train,epochs=100)
```

```python
loss, accuracy = model.evaluate(X_test_scaled,y_test)
loss, accuracy
```

```python
y_test
```

```python
predictions = np.round(model.predict(X_test_scaled))
```

```python
y_test == predictions
```

# With 2 hidden layers

```python
model2 = Sequential([])
model2.add(Dense(16, input_dim=8, activation = 'relu'))
model2.add(Dense(8, activation = 'relu'))
model2.add(Dense(1, activation = 'sigmoid'))
```

```python
model2.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
model2.fit(X_train_scaled,y_train,epochs=500)
```

```python
model2.evaluate(X_train_scaled,y_train)
```

```python
model2.evaluate(X_test_scaled,y_test)
```

```python
predictions2 = np.round(model2.predict(X_test_scaled))
predictions2 == y_test
```
